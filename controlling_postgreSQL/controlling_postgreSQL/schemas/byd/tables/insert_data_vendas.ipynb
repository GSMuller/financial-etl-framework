```python
conn.rollback()

# Ler o CSV
df = pd.read_csv(r'c:\Users\giovanni.5683\GITHUB\controlling_postgreSQL\Datasets\dta_emplacamento.csv')

# Atualizar a tabela usando UPDATE com base no idfsexterno
for index, row in df.iterrows():
    # Skip rows where dta_emplacamento is NaN
    if pd.isna(row['dta_emplacamento']):
        continue
    
    query = """
    UPDATE byd.db_vendas_byd 
    SET dta_emplacamento = %s 
    WHERE idnfsexterno = %s
    """
    cursor.execute(query, (row['dta_emplacamento'], row['idnfsexterno']))

conn.commit()
print(f"{cursor.rowcount} registros atualizados com sucesso!")
```
