```python
# Primeiro, vamos verificar o formato real da coluna competencia
cursor.execute("""
    SELECT DISTINCT competencia
    FROM byd.bonus_view
    WHERE competencia IS NOT NULL
    ORDER BY competencia DESC
    LIMIT 10
""")

amostras_competencia = cursor.fetchall()
print("Exemplos de valores na coluna 'competencia':")
for amostra in amostras_competencia:
    print(f"  - '{amostra[0]}'")

# Query ajustada para pegar todos os dezembros
query_bonus_dezembro = """
    SELECT 
        SUM(CAST(valor_bonus AS NUMERIC)) as total_valor_bonus,
        SUM(CAST(trade AS NUMERIC)) as total_trade,
        COUNT(*) as total_registros
    FROM byd.bonus_view
    WHERE competencia LIKE '%12'
       OR RIGHT(competencia, 2) = '12'
       OR competencia LIKE '____-12'
"""

cursor.execute(query_bonus_dezembro)
resultado_dezembro = cursor.fetchone()

total_valor_bonus_dezembro = resultado_dezembro[0] if resultado_dezembro[0] is not None else 0
total_trade_dezembro = resultado_dezembro[1] if resultado_dezembro[1] is not None else 0
total_registros = resultado_dezembro[2] if resultado_dezembro[2] is not None else 0

print(f"\nTotal de registros encontrados: {total_registros}")
print(f"Total Valor Bônus (Dezembro): R$ {total_valor_bonus_dezembro:,.2f}")
print(f"Total Trade (Dezembro): R$ {total_trade_dezembro:,.2f}")
```

```python
query_total_valor_bonus = """
    SELECT 
        SUM(CAST(valor_bonus AS NUMERIC)) as total_valor_bonus,
        SUM(CAST(trade AS NUMERIC)) as total_trade,
        COUNT(*) as total_carros_vendidos
    FROM byd.bonus_view
    WHERE competencia = 'Dezembro'
"""

cursor.execute(query_total_valor_bonus)
resultado_total = cursor.fetchone()

total_valor_bonus_geral = resultado_total[0] if resultado_total[0] is not None else 0
total_trade_geral = resultado_total[1] if resultado_total[1] is not None else 0
total_carros_vendidos = resultado_total[2] if resultado_total[2] is not None else 0

print(f"Total de carros vendidos em Dezembro: {total_carros_vendidos}")
print(f"Valor Total de Bônus (Dezembro): R$ {total_valor_bonus_geral:,.2f}")
print(f"Valor Total de Trade (Dezembro): R$ {total_trade_geral:,.2f}")
```
