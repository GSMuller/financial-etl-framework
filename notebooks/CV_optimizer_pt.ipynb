```python
def create_ats_cv_pdf_pt(cv_data, filename="CV_Giovanni_Muller_PT_2026.pdf"):
    """Gera PDF do curr√≠culo em portugu√™s otimizado para ATS"""
    
    output_dir = r"C:\Users\giovanni.5683\GITHUB\notebooks"
    output_path = os.path.join(output_dir, filename)
    
    doc = SimpleDocTemplate(
        output_path,
        pagesize=letter,
        rightMargin=0.75*inch,
        leftMargin=0.75*inch,
        topMargin=0.75*inch,
        bottomMargin=0.75*inch
    )
    
    styles = getSampleStyleSheet()
    
    style_name = ParagraphStyle(
        'CustomName',
        parent=styles['Heading1'],
        fontSize=18,
        textColor='black',
        spaceAfter=6,
        alignment=TA_CENTER,
        fontName='Helvetica-Bold'
    )
    
    style_contact = ParagraphStyle(
        'CustomContact',
        parent=styles['Normal'],
        fontSize=10,
        textColor='black',
        spaceAfter=3,
        alignment=TA_CENTER,
        fontName='Helvetica'
    )
    
    style_target = ParagraphStyle(
        'CustomTarget',
        parent=styles['Normal'],
        fontSize=11,
        textColor='black',
        spaceAfter=12,
        alignment=TA_CENTER,
        fontName='Helvetica-Bold'
    )
    
    style_section = ParagraphStyle(
        'CustomSection',
        parent=styles['Heading2'],
        fontSize=12,
        textColor='black',
        spaceAfter=6,
        spaceBefore=12,
        fontName='Helvetica-Bold'
    )
    
    style_job_title = ParagraphStyle(
        'CustomJobTitle',
        parent=styles['Normal'],
        fontSize=11,
        textColor='black',
        spaceAfter=2,
        fontName='Helvetica-Bold'
    )
    
    style_normal = ParagraphStyle(
        'CustomNormal',
        parent=styles['Normal'],
        fontSize=10,
        textColor='black',
        spaceAfter=3,
        fontName='Helvetica',
        leading=14
    )
    
    style_bullet = ParagraphStyle(
        'CustomBullet',
        parent=styles['Normal'],
        fontSize=10,
        textColor='black',
        spaceAfter=3,
        leftIndent=20,
        fontName='Helvetica',
        leading=14
    )
    
    story = []
    
    # HEADER
    header = cv_data['header']
    story.append(Paragraph(header['name'], style_name))
    story.append(Paragraph(header['location'], style_contact))
    story.append(Paragraph(header['target_role'], style_target))
    story.append(Paragraph(
        f"Email: {header['email']} | LinkedIn: {header['linkedin']} | GitHub: {header['github']}",
        style_contact
    ))
    story.append(Spacer(1, 0.2*inch))
    
    # RESUMO PROFISSIONAL
    story.append(Paragraph("RESUMO PROFISSIONAL", style_section))
    for para in cv_data['summary'].split('\n\n'):
        story.append(Paragraph(para.strip(), style_normal))
        story.append(Spacer(1, 0.1*inch))
    
    # COMPET√äNCIAS PRINCIPAIS
    story.append(Paragraph("COMPET√äNCIAS PRINCIPAIS", style_section))
    for category, skills in cv_data['skills'].items():
        story.append(Paragraph(f"<b>{category}:</b>", style_normal))
        for skill in skills:
            story.append(Paragraph(f"‚Ä¢ {skill}", style_bullet))
    story.append(Spacer(1, 0.1*inch))
    
    # PROJETOS
    story.append(Paragraph(cv_data['projects']['title'], style_section))
    for project in cv_data['projects']['items']:
        story.append(Paragraph(f"<b>{project['name']}</b>", style_job_title))
        story.append(Paragraph(f"GitHub: {project['github']}", style_normal))
        for desc in project['description']:
            story.append(Paragraph(f"‚Ä¢ {desc}", style_bullet))
    story.append(Spacer(1, 0.1*inch))
    
    # EXPERI√äNCIA PROFISSIONAL
    story.append(Paragraph("EXPERI√äNCIA PROFISSIONAL", style_section))
    for exp in cv_data['experience']:
        story.append(Paragraph(f"<b>{exp['title']}</b>", style_job_title))
        story.append(Paragraph(f"{exp['company']} | {exp['period']}", style_normal))
        for achievement in exp['achievements']:
            story.append(Paragraph(f"‚Ä¢ {achievement}", style_bullet))
        story.append(Spacer(1, 0.1*inch))
    
    # FORMA√á√ÉO ACAD√äMICA
    story.append(PageBreak())
    story.append(Paragraph("FORMA√á√ÉO ACAD√äMICA", style_section))
    for edu in cv_data['education']:
        story.append(Paragraph(f"<b>{edu['degree']}</b>", style_job_title))
        story.append(Paragraph(f"{edu['institution']} | {edu['period']}", style_normal))
        if 'coursework' in edu:
            story.append(Paragraph("<b>Disciplinas Relevantes:</b>", style_normal))
            for course in edu['coursework']:
                story.append(Paragraph(f"‚Ä¢ {course}", style_bullet))
        story.append(Spacer(1, 0.05*inch))
    
    # CERTIFICA√á√ïES
    story.append(Paragraph("CERTIFICA√á√ïES", style_section))
    for cert in cv_data['certifications']:
        story.append(Paragraph(f"‚Ä¢ {cert}", style_bullet))
    story.append(Spacer(1, 0.1*inch))
    
    # IDIOMAS
    story.append(Paragraph("IDIOMAS", style_section))
    for lang, level in cv_data['languages'].items():
        story.append(Paragraph(f"‚Ä¢ {lang}: {level}", style_bullet))
    story.append(Spacer(1, 0.1*inch))
    
    # INFORMA√á√ïES ADICIONAIS
    story.append(Paragraph("INFORMA√á√ïES ADICIONAIS", style_section))
    for info in cv_data['additional']:
        story.append(Paragraph(f"‚Ä¢ {info}", style_bullet))
    
    doc.build(story)
    return output_path

print("‚úÖ Fun√ß√£o para gerar PDF em portugu√™s criada!")
```
